```python
import json
import os

# os.environ["CUDA_VISIBLE_DEVICES"]="0,1,2,3"

train_json = []
valid_json = []

dir = '/home/yhjeon/projects/IITP_SE/NS_AGC/data/SITEC_AGC/train/'
for (root, directories, files) in os.walk(dir+'distorted'):
    for i, file in enumerate(files):
        if (i + 1) % 100 == 0:
                print('Processing {0} of {1}'.format(i + 1, len(files), end='\r', flush=True))
                
        
        filenum = file.split('_')[1]
        temp = {}
        path = dir+'distorted/'+ file
        temp['path'] = path
        temp['label'] = dir + 'original/original_{}.wav'.format(filenum)
        train_json.append(temp)

dir = '/home/yhjeon/projects/IITP_SE/NS_AGC/data/SITEC_AGC/valid/'
for (root, directories, files) in os.walk(dir+'distorted'):
    for i, file in enumerate(files):
        if (i + 1) % 100 == 0:
                print('Processing {0} of {1}'.format(i + 1, len(files), end='\r', flush=True))
                
        
        filenum = file.split('_')[1]
        temp = {}
        path = dir+'distorted/'+file
        temp['path'] = path
        temp['label'] = dir + 'original/original_{}.wav'.format(filenum)
        valid_json.append(temp)


       

with open('/home/yhjeon/projects/IITP_SE/NS_AGC/data/SITEC_AGC/train.json', 'w') as out:
    json.dump(train_json, out, indent=4)
with open('/home/yhjeon/projects/IITP_SE/NS_AGC/data/SITEC_AGC/valid.json', 'w') as out:
    json.dump(valid_json, out, indent=4)
```
