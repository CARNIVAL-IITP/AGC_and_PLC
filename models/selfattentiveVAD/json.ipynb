```python
dir = '/home/yhjeon/dns_db/clean/'

train_json = []
valid_json = []
test_json = []
for (root, directories, files) in os.walk(dir):
    for i, file in enumerate(files):
        if i % 100 == 0:
                print('Processing {0} of {1}'.format(i + 1, len(files), end='\r', flush=True))
        
        file_num = int(file.split('_')[-1].split('.')[0])
        if file_num >= 6000:
            continue
        temp = {}
        path = root + file
        sample, sr = sf.read(path)
        temp['path'] = path
        temp['length'] = sample.shape[0]

        output = pipeline("/home/yhjeon/dns_db/clean/clean_fileid_" + str(file_num) + '.wav')
        timeline = output.get_timeline()
        temp['label'] = []
        for time in timeline:
            time_start = int(float(str(time).split(' --> ')[0][8:]) * sr)
            time_end = int(float(str(time).split(' --> ')[1][7:-1]) * sr)
            temp['label'].append([time_start, time_end])

        if file_num < 5000:
            train_json.append(temp)
        elif file_num < 5500:
            valid_json.append(temp)
        elif file_num < 6000:
            test_json.append(temp)
        

""" with open('/home/yhjeon/AGC/src/vad/train_clean.json', 'w') as out:
    json.dump(train_json, out, indent=4)
with open('/home/yhjeon/AGC/src/vad/valid_clean.json', 'w') as out:
    json.dump(valid_json, out, indent=4)
with open('/home/yhjeon/AGC/src/vad/test_clean.json', 'w') as out:
    json.dump(test_json, out, indent=4) """
```
